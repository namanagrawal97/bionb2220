# Importing necessary python modules and our data

```python
import sklearn
from sklearn import datasets

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%pip install ipywidgets
```

```python
iris=datasets.load_iris()
x=pd.DataFrame(iris['data'],columns=iris['feature_names'])
y=pd.DataFrame(iris['target'],columns=['target'])
iris_data=pd.concat([x,y],axis=1)
iris_data.head(3)
```

Each row represents one flower.
There are 4 **features** of every flower - sepal length, sepal width, petal length and petal width.

The column **'target'** is the **true identity of the flower**. It is either 0, i.e. the flower is an *Iris setosa*,
1- the flower is an *Iris versicolor*
2 - the flower is an *Iris virginica*
This column is called 'Target', because we want to build an AI that can predict which species the flower is of based on the features of the flower.

Generally, features become correlated. In simpler terms, a flower that has smaller petals might also have smaller sepals. Another species might have larger sepals correlated with larger petals. In other words, we want our AI to identify patterns and separate out the species based on it.classifier.jpg

###Great! it seems like our data should be easily separable by drawing a line. Now lets try to ge ta machine to do draw the line for us :)

* *for here, we need to let students which 2 flowers they want to separate, and based on which two features do they want to separate*

```python
#preparing our data to be understood by our model
X = iris_data.iloc[0:100, [2,3]].values
y_data = iris_data.iloc[0:100, 4].values
y_data = np.where(y_data == 0, -1, 1)

from sklearn.linear_model import Perceptron


ppn = Perceptron(tol=1e-3, random_state=0) #initializing a new perceptron
ppn.fit(X, y_data) #training
```

Our model is now trained ! Phew that was quick. Lets see how our model performed
* *for here, we need to insert some explanation of different validation scores*

```python
from sklearn.metrics import classification_report

# Use the same data for prediction
y_pred = ppn.predict(X)

# Print the classification report
print(classification_report(y_data, y_pred))

# Print the accuracy
print("Accuracy: ", ppn.score(X, y_data))
```

Lets get a visual representation of how our model is performing
* *here, we need to make sure that the species and the features that the students have chosen get accurately represented*

```python
# Get the coefficients and intercept
coef = ppn.coef_[0]
intercept = ppn.intercept_

# Function to plot decision boundary
def plot_decision_boundary(X, y, coef, intercept):
    plt.figure(figsize=(10, 6))

    # Plotting the data points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='Paired', edgecolor='k')

    # Calculate the decision boundary line
    xmin, xmax = plt.xlim()
    ymin, ymax = plt.ylim()
    x_values = np.array([xmin, xmax])
    y_values = -(intercept + coef[0] * x_values) / coef[1]

    # Plotting the decision boundary line
    plt.plot(x_values, y_values, 'k--')

    plt.xlabel('Petal Width')
    plt.ylabel('Petal Length')
    plt.title('Decision Boundary of a Trained Perceptron Model')
    plt.show()

# Plot the decision boundary
plot_decision_boundary(X, y_data, coef, intercept)
```

Let's see how our model makes the decision over time as we train it

```python
import imageio
import matplotlib.pyplot as plt

# Train the Perceptron and store the number of errors for each epoch
num_epochs = 10
errors_per_epoch = []
images = []
for epoch in range(num_epochs):
    ppn.partial_fit(X, y_data, classes=np.unique(y_data))
    errors_per_epoch.append(sum(y_data != ppn.predict(X)))

    # Plot the decision boundary
    plot_decision_boundary(X, y_data, ppn.coef_[0], ppn.intercept_)

    # Save the current figure to a file
    filename = f'plot_{epoch}.png'
    plt.savefig(filename)
    plt.close()  # Close the figure to free up memory
    images.append(imageio.imread(filename))

# Create a GIF
imageio.mimsave('training.gif', images, fps=3)
```
